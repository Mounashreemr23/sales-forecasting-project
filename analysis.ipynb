```python
monthly_df = monthly_sales.reset_index()
monthly_df.columns = ['Month', 'Sales']
monthly_df['Month'] = pd.to_datetime(monthly_df['Month'])

monthly_df['Year'] = monthly_df['Month'].dt.year
monthly_df['Month_Num'] = monthly_df['Month'].dt.month
monthly_df['Quarter'] = monthly_df['Month'].dt.quarter

monthly_df['Lag_1'] = monthly_df['Sales'].shift(1)
monthly_df['Lag_2'] = monthly_df['Sales'].shift(2)
monthly_df['Lag_3'] = monthly_df['Sales'].shift(3)

monthly_df.dropna(inplace=True)

monthly_df.to_csv('data/monthly_sales_features.csv', index=False)

print("Feature engineering completed successfully!")
monthly_df.head()
```

```python
# Input features
X = monthly_df[['Year', 'Month_Num', 'Quarter', 'Lag_1', 'Lag_2', 'Lag_3']]

# Target variable
y = monthly_df['Sales']

print(X.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False
)

print("Training rows:", len(X_train))
print("Testing rows:", len(X_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
y_pred = model.predict(X_test)

print(y_pred[:5])
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE :", round(mae, 2))
print("RMSE:", round(rmse, 2))
print("R²  :", round(r2, 4))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(y_test.values, label='Actual Sales')
plt.plot(y_pred, label='Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Test Months')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import joblib

joblib.dump(model, 'models/sales_forecast_model.pkl')

print("Model saved successfully!")
```

```python
# Last row of historical data
last_row = monthly_df.iloc[-1]

print(last_row)
```

```python
future_predictions = []

# Start with the latest known row
current_row = last_row.copy()

for i in range(12):
    # Move to next month
    next_month = current_row['Month'] + pd.DateOffset(months=1)

    year = next_month.year
    month_num = next_month.month
    quarter = next_month.quarter

    # Prepare input for model
    input_data = pd.DataFrame([{
        'Year': year,
        'Month_Num': month_num,
        'Quarter': quarter,
        'Lag_1': current_row['Sales'],
        'Lag_2': current_row['Lag_1'],
        'Lag_3': current_row['Lag_2']
    }])

    # Predict sales
    predicted_sales = model.predict(input_data)[0]

    # Save result
    future_predictions.append({
        'Month': next_month,
        'Predicted Sales': predicted_sales
    })

    # Update current row for next iteration
    current_row = {
        'Month': next_month,
        'Sales': predicted_sales,
        'Lag_1': current_row['Sales'],
        'Lag_2': current_row['Lag_1']
    }

# Convert to DataFrame
future_df = pd.DataFrame(future_predictions)

future_df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(future_df['Month'], future_df['Predicted Sales'], marker='o')
plt.title('12-Month Sales Forecast')
plt.xlabel('Month')
plt.ylabel('Predicted Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
future_df.to_csv('data/future_sales_forecast.csv', index=False)
print("Forecast saved successfully!")
```
